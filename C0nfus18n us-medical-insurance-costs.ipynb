```python
# First Import the csv and check
import csv

# We use with to open the csv, then we use DictReader to make of each line a dictionary
# insurance_data as list comprehension -> we append each row of insurance reader dictionary to the list insurance data.

with open("insurance.csv") as insurance_doc:
    insurance_reader = csv.DictReader(insurance_doc)
    insurance_data = [row for row in insurance_reader] 

print(insurance_data)
```

```python
# We create a function to count the total patients

def count_patients(data):
    total_patients = 0
    for row in data:
        total_patients += 1
        
    return total_patients
    
```

```python
print("we have a total of " + str(count_patients(insurance_data)) + " patients")
```

```python
def count_sex(data):
    total_patients = count_patients(insurance_data)
    total_male_patients = 0
    
    for i in data:
        if i["sex"] == "male":
            total_male_patients += 1
        else:
            total_male_patients += 0
            
    total_female_patients = total_patients - total_male_patients
    
    return print("Male patients = " + str(total_male_patients) + "\n Female patients =" + str(total_female_patients))
```

```python
count_sex(insurance_data)
```

```python
def count_smoker(data):
    total_patients = count_patients(insurance_data)
    total_smokers = 0
    
    for i in data:
        if i["smoker"] == "yes":
            total_smokers += 1
        else:
            total_smokers += 0
            
    total_non_smokers = total_patients - total_smokers
    
    return print("Smoker patients = " + str(total_smokers) + "\n Female patients =" + str(total_non_smokers))
```

```python
count_smoker(insurance_data)
```

```python
#TO CHECK THE TYPE OF DATA THE VALUE HAS

def check_value_types(data):
    for row in data:
        for key, value in row.items():
            print(f"The key '{key}' has a value '{value}' of type {type(value)}")

# Llama a la funcin para verificar los tipos de valores
check_value_types(insurance_data)
```

```python
def average_insurance_cost(data):
    total_cost = 0
    number_of_patients = count_patients(insurance_data)
    
    for row in data:
        charges = float(row["charges"])
        total_cost += charges
    
    return round((total_cost / number_of_patients),2)
```

```python
average_insurance_cost(insurance_data)
```

```python
def max_min_insurance_cost(data):
    charges_list = [float(row["charges"]) for row in data]
    
    return max(charges_list), min(charges_list)
    
```

```python
max_min_insurance_cost(insurance_data)
```

```python
regions =[]

for row in insurance_data:
    if row["region"] not in regions:
        regions.append(row["region"])

print(regions)
```

```python
def expensives_regions(data):
    southwest_total_charges = 0
    southwest_total_patients = 0
    
    southeast_total_charges = 0
    southeast_total_patients = 0
    
    
    northwest_total_charges = 0
    northwest_total_patients = 0
    
    northeast_total_charges = 0
    northeast_total_patients = 0
    
    for row in data:
        if row["region"] == "southwest":
            southwest_total_charges += float(row["charges"])
            southwest_total_patients += 1
        
        elif row["region"] == "southeast":
            southeast_total_charges += float(row["charges"])
            southeast_total_patients += 1
        
        elif row["region"] == "northwest":
            northwest_total_charges += float(row["charges"])
            northwest_total_patients += 1
           
        elif row["region"] == "northeast":
            northeast_total_charges += float(row["charges"])
            northeast_total_patients += 1
    
    average_southwest = southwest_total_charges / southwest_total_patients
    average_southeast = southeast_total_charges / southeast_total_patients
    average_northwest = northwest_total_charges / northwest_total_patients
    average_northeast = northeast_total_charges / northeast_total_patients
    
    
    Average_region = [average_southwest, average_southeast, average_northwest, average_northeast]
    most_expensive = max(Average_region)
    most_cheap = min(Average_region)
    print(most_expensive)
    print(most_cheap)
    
    average_per_region = {average_southwest : "Southwest", average_southeast: "Southeast", average_northwest: "Northwest", average_northeast : "Northeast"}

    return print("The most expensive region is: " + str(average_per_region[most_expensive]) + " with an average cost of " + str(most_expensive)),print( "The cheapest region is : " + str(average_per_region[most_cheap]) + " with an average cost of "+ str(most_cheap))
    


```

```python
expensives_regions(insurance_data)
```
